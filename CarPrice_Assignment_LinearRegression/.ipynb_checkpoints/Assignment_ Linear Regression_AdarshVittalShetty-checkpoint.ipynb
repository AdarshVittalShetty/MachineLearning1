## Step 3: Preparing the Data

```python
#Preparing the categorical Data

#Function defined to get dummy variables for each categorical data, drop the first dummy column and original column
#return concaneted dataframe
def getDummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    #drop the original column
    df.drop([x], axis = 1, inplace = True)
    return df

# Applying the function to the categorical data of the dataframe
df = getDummies('fueltype',df)
df = getDummies('aspiration',df)
df = getDummies('carbody',df)
df = getDummies('drivewheel',df)
df = getDummies('enginetype',df)
df = getDummies('cylindernumber',df)
df = getDummies('doornumber',df)
df = getDummies('fuelsystem',df)
df = getDummies('enginelocation',df)

# we do not need company name and car ID for our price analysis
df.drop(['company','car_ID'],axis = 1, inplace = True)
```

## Step 4: Splitting the Data into Training and Testing Sets

```python
# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7,test_size = 0.3, random_state = 100)
```

```python
#Normalizing the values using MinMaxScaler
scaler = MinMaxScaler()
columns = df_train.columns
df_train[columns] = scaler.fit_transform(df_train[columns])
```

## Step 5: Building a linear model :RFE

```python
# Dividing into X and Y set
y_train = df_train.pop('price')
X_train = df_train
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#Top 10 columns shorlisted by RFE process
col = X_train.columns[rfe.support_]
col
```

#### Model 1 : 
- Using All the columns obtained after Recurssive Feature Elimination

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
# Adding a constant variable 
X_train_lm_1 = sm.add_constant(X_train_rfe)
# fitting a linear model
lm = sm.OLS(y_train,X_train_lm_1).fit()
#summary
print(lm.summary())
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Reusable Function to drop a column and create a model
def createModel(X_train_rfe,dropColumn):
    X_train_rfe = X_train_rfe.drop(dropColumn, axis = 1)
    # Adding a constant variable 
    X_train_new = sm.add_constant(X_train_rfe)
    # fitting a linear model
    lm = sm.OLS(y_train,X_train_new).fit()
    return lm,X_train_rfe,X_train_new

#Reusable function to show VIF value
def showVIF(X_train_rfe):
    # Create a dataframe that will contain the names of all the feature variables and their respective VIFs
    vif = pd.DataFrame()
    vif['Features'] = X_train_rfe.columns
    vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    print(vif)
```

#### Model 2 
- All the independent variables are significant, curbweight has VIF>5
- Let us drop curbweight and build a model

```python
#Create Model
lm,X_train_rfe,X_train_new = createModel(X_train_rfe,['curbweight'])

#Print model summary
print(lm.summary())

#Display the Variance Inflation Factor
showVIF(X_train_rfe)
```

#### Model 3

- `horsepower` has some insignifance and `VIF > 5`
- Let us drop horsepower and build the model

```python
#Create Model
lm,X_train_rfe,X_train_new = createModel(X_train_rfe,['horsepower'])

#Print model summary
print(lm.summary())

#Display the Variance Inflation Factor
showVIF(X_train_rfe)
```
